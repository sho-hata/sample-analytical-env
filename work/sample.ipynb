```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
male = np.array([173, 170, 180, 177, 175, 176, 173, 181, 174, 169])
female = np.array([153, 150, 154, 157, 151, 159, 151, 153, 154, 154])
```

```python
# Compare histograms
fig = plt.figure()
ax = fig.add_subplot(111)
ax.hist(male, bins=15,  color='blue', alpha = 0.7)
ax.hist(female, bins=15, color='red', alpha = 0.7)
ax.set_title('male:Blue   female:Female')
ax.set_xlabel('height')
ax.set_ylabel('proportion')
fig.show()
```

```python
# unbiased sample variance
male_var = np.var(male, ddof=1)
female_var = np.var(female, ddof=1)
# degree of freedom
male_df = len(male) - 1
female_df = len(female) - 1
# 
f = male_var / female_var  # F比の値
# one-sided	test
one_sided_pval1 = stats.f.cdf(f, male_df, female_df) 
one_sided_pval2 = stats.f.sf(f, male_df, female_df)
# two-sided	test
two_sided_pval = min(one_sided_pval1, one_sided_pval2) * 2

print('F:       ', round(f, 3))
print('p-value: ', round(two_sided_pval, 4))
```
